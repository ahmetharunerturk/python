```python
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['AAPL'])
# Interest by Region
df = pytrend.interest_by_region()
df.head(10)
```

```python
df.reset_index().plot(x='geoName', y='AAPL', figsize=(120, 10), kind ='bar')
```

```python
df = pytrend.trending_searches(pn='united_states')
df.head()
```

```python
df = pytrend.today_searches(pn='US')
```

```python
df = pytrend.top_charts(2019, hl='en-US', tz=300, geo='GLOBAL')
df.head()
```

```python
keywords = pytrend.suggestions(keyword='Mercedes Benz')
df = pd.DataFrame(keywords)
df.drop(columns= 'mid')
```
